## Highest valuation ask vs final valuation

```python
df2 = df[df['deal'] == 1 ][['ask_valuation','deal_valuation']].sort_values(by = 'ask_valuation',ascending = False)
df2
```

```python
plt.scatter(x = df2['ask_valuation'], y = df2['deal_valuation'])
plt.title(' valuation ask vs final valuation')
plt.xlabel('ask_valuation')
plt.ylabel('deal_valuation');
```

## Industry/category wise success rate

```python
df['category'] = 'Other'

df.loc[df['idea'].str.contains('food|snack|drink|momos|chips|ice cream|sweets|kheer|oil|egg|milk|lemonade', case=False, na=False), 'category'] = 'Food'

df.loc[df['idea'].str.contains('app|software|ai|tech|device|smart|platform|detector|machine|vr', case=False, na=False), 'category'] = 'Tech'

df.loc[df['idea'].str.contains('wear|clothing|underwear|fashion|streetwear|bags|shoes', case=False, na=False), 'category'] = 'Fashion'

df.loc[df['idea'].str.contains('health|medical|ecg|diabetes|dialysis|hygiene', case=False, na=False), 'category'] = 'Health'
```

```python
df['category'].value_counts()
```

```python
pd.crosstab(df['category'], df['deal'])
```

```python
data = (df.groupby('category')['deal'].mean()*100)
data
```

```python
plt.bar(data.index, data.values)
plt.ylabel('Percentag');
```

## Episode wise funding trend

```python
df.groupby('episode_number')['deal_amount'].sum()
```

```python
df2 = df.groupby('episode_number')['deal_amount'].sum()
plt.bar(df2.index,df2.values)
plt.xlabel('ep')
plt.ylabel('amt');
```

## Shark wise investment count

```python
data = df[['anupam_deal','aman_deal','namita_deal','vineeta_deal','peyush_deal','ghazal_deal']].sum()
data
```

```python
plt.bar(data.index,data.values)
plt.title('Shark wise investment count')
plt.xlabel('shark name')
plt.ylabel('deal count')
plt.xticks(rotation  = 60);
```

## Shark wise total amount invested

```python
df.loc[df['anupam_deal'] == 1 ,'amount_per_shark'].sum()
```

```python
sharks = ['ashneer','anupam','aman','namita','vineeta','peyush','ghazal']
totals = []

for shark in sharks:
    total = df.loc[df[f'{shark}_deal'] == 1, 'amount_per_shark'].sum()
    totals.append(total)
    print(shark, total)
```

```python
pd.Series(totals, index=sharks).plot(kind='bar')
plt.ylabel('total amount');
```
