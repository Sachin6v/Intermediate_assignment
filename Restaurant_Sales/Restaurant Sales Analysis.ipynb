```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#   ---------- Restaurant Sales Analysis using Python ---------------

```python
df = sns.load_dataset('tips')
df.head()
```

# Total revenue kitna hai?
(total_bill sum)

```python
df['total_bill'].sum()
```

```python
plt.hist(x = df['total_bill']);
plt.xlabel('bill');
```

# Average tip kitni hai?

```python
df['tip'].mean()
```

 # Kis day sabse zyada sales hui?

```python
df.groupby(['day'])['total_bill'].sum().sort_values(ascending = False).reset_index()
```

```python
data = df.groupby(['day'])['total_bill'].sum()
```

```python
plt.bar(data.index, data.values)
plt.ylabel('bill');
```

# Lunch ya Dinner me zyada revenue?

```python
df.groupby('time')['total_bill'].sum().sort_values(ascending = False).reset_index()
```

```python
data = df.groupby('time')['total_bill'].sum()
```

```python
plt.bar(data.index, data.values)
plt.xlabel('time')
plt.ylabel('bill');
```

# Male ya Female customers zyada aaye?

```python
df['sex'].value_counts()
```

```python
data = df['sex'].value_counts()
```

```python
plt.bar(data.index, data.values)
plt.xlabel('customers')
plt.ylabel('count');
```

# Kis day average tip sabse zyada hai?

```python
df.groupby('day')['tip'].sum().sort_values(ascending = False).reset_index()
```

```python
data = df.groupby('day')['tip'].sum().sort_values(ascending = False).reset_index()
plt.bar(x = data['day'], height = data['tip'] )
plt.xlabel('day')
plt.ylabel('tip');
```

# Smokers zyada tip dete hain ya non-smokers?

```python
df.groupby('smoker')['tip'].sum()
```

```python
data  = df.groupby('smoker')['tip'].sum()
plt.bar(data.index, data.values)
plt.xlabel("Smoker")
plt.ylabel('tip');
```

# Top 5 highest bills dikhao

```python
df.nlargest(5,'total_bill')
```

# Kis time average bill zyada hai?

```python
df.groupby('time')['total_bill'].mean()
```

```python
data = df.groupby('time')['total_bill'].mean()
```

```python
plt.pie(data.values,
        labels=data.index,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Lunch vs Dinner Revenue Share');
```
